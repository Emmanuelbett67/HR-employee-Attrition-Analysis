### Introduction

- <b>Employee Attrition</b> is also known as employee turnover, refers to the process by which employees leave an organization and are subsequently replaced by new employees. This can occur through various forms such as resignations, retirements, layoffs, or terminations. Understanding and managing employee attrition is crucial for organizations because high attrition rates can lead to increased costs, loss of institutional knowledge, and decreased morale among remaining employees.

## Exploratory Data Analysis
*Identifying trends and patterns in the data*
### 1. Employee Attrition Rate

```python
#viz to show employee attrition rate
#value_counts shows the number of occurrence of yes and no
attrition_rate = df["Attrition"].value_counts()
labels = ["No", "Yes"]

#percent calculation
total = sum(attrition_rate)
pct = [(count/total)*100 for count in attrition_rate]

#bar plot
fig, ax = plt.subplots()
bars = ax.bar(labels, attrition_rate, color=["#344F9E","#E42D2E"])
    
#axis labels, title and grid
ax.set_xlabel("Attrition")
ax.set_ylabel("Number of employees")
ax.set_title("Attrition Rate")

#percent on top of bars
for bar, count, percent in zip(bars, attrition_rate, pct):
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f'{int(yval)} ({percent:.1f}%)', va='bottom', ha='center')
```

*Point to note:*
- The employee attrition rate in this organization is 16.1%. This is considered high as experts in Human Resource suggest tha a normal attrition rate for organizations is between 4% to 7%.

### 2. Employee Attrition by Age

```python
#viz
sns.histplot(x="Age",hue="Attrition",data=df,kde=True,palette=["#11264e","skyblue"])
plt.title("Employee Distribution by Age",fontweight="black",size=20,pad=10)
plt.show()
```

*Points to note:*
- In this organization most of the employees are between the age of 30 to 40.
- Employees with a young age leaves the company more than the elder employees.
- There is a clear trend that **Attrition** decreases as **Age** increases.

### 5. Employee Attrition by Department


```python
# Calculate the total number of employees in each department
total_by_department = df['Department'].value_counts()

# Calculate the number of employees who have left for each department
attrition_by_department = df[df['Attrition'] == 'Yes']['Department'].value_counts()

# Calculate the attrition rate for each department
attrition_rate_by_department = (attrition_by_department / total_by_department) * 100

#sort in descending order
attrition_rate_by_department = attrition_rate_by_department.sort_values(ascending=False)

# Plotting
bars = plt.bar(attrition_rate_by_department.index, attrition_rate_by_department.values, color=['#344F9E', '#E42D2E', '#D0A153'])

# Adding labels and title
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.title('Employee Attrition Rate by Department')

# Adding percentages on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, f'{yval:.1f}%', ha='center', va='bottom')

plt.show()
```

*Point to note:*
- The highest attrition rate is in the **Sales Department**, followed closely by the Human **Resource department**

### 6. Employee Attrition by Education


```python
# total number of employees for each education level
total_by_education = df['Education'].value_counts()

# Calculate the number of employees who have left for each education level
attrition_by_education = df[df['Attrition'] == 'Yes']['Education'].value_counts()

#Calculate the attrition rate for each education level
attrition_rate_by_education = (attrition_by_education / total_by_education) * 100

# Sort the attrition rates in descending order
attrition_rate_by_education = attrition_rate_by_education.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(attrition_rate_by_education.index, attrition_rate_by_education.values, color=['black', '#344F9E', '#E42D2E', '#D0A153', 'skyblue'])

# Adding labels and title
plt.xlabel('Education Level')
plt.ylabel('Attrition Rate (%)')
plt.title('Employee Attrition Rate by Education Level')

# Adding percentages on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, f'{yval:.1f}%', ha='center', va='bottom')

plt.show()
```

### 9. Employee Attrition by Job Role and Monthly Income


```python
# Create income bins
income_bins = [0, 2000, 4000, 6000, 8000, 10000, 12000, 14000, 16000, 18000, 20000]
df['IncomeRange'] = pd.cut(df['MonthlyIncome'], bins=income_bins)

# Pivot table to calculate attrition rates
pivot_table = pd.pivot_table(df, values='Attrition', index='JobRole', columns='IncomeRange', aggfunc=lambda x: (x == 'Yes').mean() * 100)

# Plotting
plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlGnBu", cbar_kws={'label': 'Attrition Rate (%)'})

# Adding labels and title
plt.xlabel('Monthly Income Range')
plt.ylabel('Job Role')
plt.title('Employee Attrition Rate by Job Role and Monthly Income Range')

plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

*Point to note:*
- **Laboratory Technicians**, **Research Scientists** and **Sales Representatives** have very low salary and this could be a major factor behind attritions.
